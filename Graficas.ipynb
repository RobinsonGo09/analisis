```python
import matplotlib.pyplot as plt  # Importar la biblioteca Matplotlib y su módulo pyplot para generar gráficos

%matplotlib inline  # Permite mostrar los gráficos directamente en Jupyter Notebook o Google Colab
```

```python
x=[1,2,3,4,5]  # Lista de valores en el eje X (tiempo en meses)
y=[10,13,15,20,25]  # Lista de valores en el eje Y (valores correspondientes a cada mes)

# Crear el gráfico
plt.plot(x,y, marker="o", linestyle="-", color ="b", label="Crecimiento")  
# marker="o" → Dibuja un marcador circular en cada punto
# linestyle="-" → Usa una línea continua
# color="b" → Color azul
# label="Crecimiento" → Etiqueta que aparecerá en la leyenda

plt.xlabel("Tiempo (meses)")  # Etiqueta del eje X
plt.ylabel("Valor")  # Etiqueta del eje Y
plt.title("Ejemplo grafica de linea")  # Título del gráfico
plt.legend()  # Muestra la leyenda con la etiqueta definida en "label"
plt.show()  # Muestra el gráfico en pantalla
```

```python
import numpy as np  # Importar la biblioteca NumPy para trabajar con arreglos numéricos

x = np.linspace(0, 10, 100)  # Generar 100 valores equidistantes entre 0 y 10

y1 = np.sin(x)  # Calcular el seno de cada valor en x
y2 = np.cos(x)  # Calcular el coseno de cada valor en x

# Definir el tamaño de la gráfica
plt.figure(figsize=(8,5))  # Establecer un tamaño de 8x5 pulgadas para la figura

# Graficar las funciones seno y coseno
plt.plot(x, y1, label="Seno", color="r", linestyle="--")  # Graficar la función seno en rojo con línea discontinua
plt.plot(x, y2, label="Coseno", color="g", linestyle="-.")  # Graficar la función coseno en verde con línea de puntos y guiones

plt.xlabel("x - Valores")  # Etiqueta del eje X
plt.ylabel("y - Función")  # Etiqueta del eje Y
plt.title("Funciones de Seno y Coseno")  # Título de la gráfica
plt.legend()  # Mostrar la leyenda con las etiquetas definidas en "label"
plt.grid(True)  # Mostrar una cuadrícula en el fondo para facilitar la lectura
plt.show()  # Mostrar la gráfica en pantalla
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lista de nombres de los productos
productos = ["Producto A", "Producto B", "Producto C"]

# Lista de etiquetas para los trimestres
trimestre = ["T1", "T2", "T3", "T4"]

# Matriz de ventas de cada producto en cada trimestre
ventas = np.array([[150, 180, 220, 260],  # Ventas del Producto A
                   [200, 210, 250, 290],  # Ventas del Producto B
                   [170, 190, 230, 250]   # Ventas del Producto C
                  ])

x = np.arange(len(trimestre))  # Genera una secuencia numérica para las posiciones de los trimestres en el eje X
width = 0.3  # Define el ancho de las barras

# Crear la figura y los ejes del gráfico con un tamaño de 8x6 pulgadas
fig, ax = plt.subplots(figsize=(8,6))

# Graficar las barras para cada producto con desplazamiento para que no se sobrepongan
ax.bar(x - width, ventas[0], width, label=productos[0], color="b")  # Producto A en azul, desplazado a la izquierda
ax.bar(x, ventas[1], width, label=productos[1], color="r")  # Producto B en rojo, centrado
ax.bar(x + width, ventas[2], width, label=productos[2], color="g")  # Producto C en verde, desplazado a la derecha

# Etiquetas de los ejes y título del gráfico
ax.set_xlabel("Trimestre")  # Etiqueta del eje X
ax.set_ylabel("Ventas")  # Etiqueta del eje Y
ax.set_title("Comparación de ventas por trimestre")  # Título del gráfico

# Configurar las etiquetas del eje X
ax.set_xticks(x)  # Definir la ubicación de los ticks en el eje X
ax.set_xticklabels(trimestre)  # Asignar los nombres de los trimestres a los ticks

ax.legend()  # Mostrar la leyenda con los nombres de los productos

# Agregar una cuadrícula en el eje Y con líneas punteadas y algo de transparencia
ax.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()  # Mostrar el gráfico
```
