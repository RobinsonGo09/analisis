```python
# Agrupar por país sumando el consumo total de energía
df_2020 = df_2020.groupby("Country", as_index=False)["Total Energy Consumption (TWh)"].sum()

# Generar una lista de colores únicos para cada país
colormap = plt.colormaps["tab10"]
num_paises = len(df_2020)

# Crear la gráfica
plt.figure(figsize=(12, 6))
plt.bar(df_2020["Country"], df_2020["Total Energy Consumption (TWh)"], 
        color=[colormap(i % 10) for i in range(num_paises)], edgecolor='black')

# Personalizar la gráfica
plt.xlabel("País")
plt.ylabel("Consumo Total de Energía (TWh)")
plt.title("Consumo Total de Energía por País en 2020")
plt.xticks(rotation=90)  # Rotar nombres de los países para mejor visualización

# Mostrar la gráfica
plt.show()
```

```python
# Filtrar los datos solo para Canadá
df_canada = df[df["Country"] == "Canada"]

# Agrupar por año y sumar el consumo per cápita
df_canada_grouped = df_canada.groupby("Year", as_index=False)["Per Capita Energy Use (kWh)"].sum()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 5))
plt.plot(df_canada_grouped["Year"], df_canada_grouped["Per Capita Energy Use (kWh)"], marker='o', linestyle='-', color='b', linewidth=2, markersize=6, label="Consumo Per Cápita")

# Personalizar el gráfico
plt.xlabel("Año", fontsize=12)
plt.ylabel("Consumo Per Cápita de Energía (kWh)", fontsize=12)
plt.title("Evolución del Consumo Per Cápita de Energía en Canadá (2020-2025)", fontsize=14)
plt.xticks(df_canada_grouped["Year"], rotation=45)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gráfico
plt.show()

```

```python
# Filtrar los datos para el año 2023
df_2023 = df[df["Year"] == 2023]

# Seleccionar solo las columnas necesarias
df_2023 = df_2023.groupby("Country", as_index=False)["Renewable Energy Share (%)"].mean()

# Crear el gráfico circular
plt.figure(figsize=(10, 8))
plt.pie(df_2023["Renewable Energy Share (%)"], labels=df_2023["Country"], autopct="%1.1f%%", 
        colors=plt.cm.tab20.colors, startangle=140)
plt.title("Porcentaje de Energía Renovable por País en 2023")
plt.axis("equal")  # Para mantener la proporción del círculo
plt.show()


```

```python
# Filtrar los datos solo para Canadá y Alemania
df_canada = df[df["Country"] == "Canada"]
df_germany = df[df["Country"] == "Germany"]

# Agrupar por año y sumar el consumo per cápita
df_canada_grouped = df_canada.groupby("Year", as_index=False)["Per Capita Energy Use (kWh)"].sum()
df_germany_grouped = df_germany.groupby("Year", as_index=False)["Per Capita Energy Use (kWh)"].sum()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 5))
plt.plot(df_canada_grouped["Year"], df_canada_grouped["Per Capita Energy Use (kWh)"], marker='o', linestyle='-', color='b', linewidth=2, markersize=6, label="Canadá")
plt.plot(df_germany_grouped["Year"], df_germany_grouped["Per Capita Energy Use (kWh)"], marker='s', linestyle='--', color='r', linewidth=2, markersize=6, label="Alemania")

# Personalizar el gráfico
plt.xlabel("Año", fontsize=12)
plt.ylabel("Consumo Per Cápita de Energía (kWh)", fontsize=12)
plt.title("Evolución del Consumo Per Cápita de Energía (2020-2025)", fontsize=14)
plt.xticks(df_canada_grouped["Year"], rotation=45)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar datos solo para Rusia
df_rusia = df[df["Country"] == "Russia"]

# Agrupar por año y calcular el promedio
df_rusia_grouped = df_rusia.groupby("Year", as_index=False).agg({
    "Renewable Energy Share (%)": "mean",  # Promedio de energía renovable por año
    "Carbon Emissions (Million Tons)": "sum"  # Suma total de emisiones por año
})

# Crear la figura
plt.figure(figsize=(8, 5))

# Gráfico de dispersión con datos agrupados
plt.scatter(df_rusia_grouped["Renewable Energy Share (%)"], df_rusia_grouped["Carbon Emissions (Million Tons)"], 
            color="blue", alpha=0.6, label="Datos Promediados por Año")

# Ajustar línea de tendencia (Regresión lineal)
coef = np.polyfit(df_rusia_grouped["Renewable Energy Share (%)"], df_rusia_grouped["Carbon Emissions (Million Tons)"], 1)
tendencia = np.poly1d(coef)

# Dibujar la línea de tendencia
x_tendencia = np.linspace(df_rusia_grouped["Renewable Energy Share (%)"].min(), 
                          df_rusia_grouped["Renewable Energy Share (%)"].max(), 100)
y_tendencia = tendencia(x_tendencia)
plt.plot(x_tendencia, y_tendencia, color="red", linestyle="--", label="Tendencia")

# Personalizar gráfico
plt.xlabel("Porcentaje de Energía Renovable (%)", fontsize=12)
plt.ylabel("Emisiones de Carbono (Millones de Toneladas)", fontsize=12)
plt.title("Relación entre Energía Renovable y Emisiones - Rusia (Promedio por Año)", fontsize=14)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Mostrar gráfico
plt.show()


```
