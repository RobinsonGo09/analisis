```python
# Filtrar los datos solo para Canadá
df_canada = df[df["Country"] == "Canada"]

# Agrupar por año y sumar el consumo per cápita
df_canada_grouped = df_canada.groupby("Year", as_index=False)["Per Capita Energy Use (kWh)"].sum()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 5))
plt.plot(df_canada_grouped["Year"], df_canada_grouped["Per Capita Energy Use (kWh)"], marker='o', linestyle='-', color='b', linewidth=2, markersize=6, label="Consumo Per Cápita")

# Personalizar el gráfico
plt.xlabel("Año", fontsize=12)
plt.ylabel("Consumo Per Cápita de Energía (kWh)", fontsize=12)
plt.title("Evolución del Consumo Per Cápita de Energía en Canadá (2020-2025)", fontsize=14)
plt.xticks(df_canada_grouped["Year"], rotation=45)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gráfico
plt.show()

```

```python
# Filtrar los datos para el año 2023
df_2023 = df[df["Year"] == 2023]

# Seleccionar solo las columnas necesarias
df_2023 = df_2023.groupby("Country", as_index=False)["Renewable Energy Share (%)"].mean()

# Crear el gráfico circular
plt.figure(figsize=(10, 8))
plt.pie(df_2023["Renewable Energy Share (%)"], labels=df_2023["Country"], autopct="%1.1f%%", 
        colors=plt.cm.tab20.colors, startangle=140)
plt.title("Porcentaje de Energía Renovable por País en 2023")
plt.axis("equal")  # Para mantener la proporción del círculo
plt.show()


```

```python
# Filtrar los datos solo para Canadá y Alemania
df_canada = df[df["Country"] == "Canada"]
df_germany = df[df["Country"] == "Germany"]

# Agrupar por año y sumar el consumo per cápita
df_canada_grouped = df_canada.groupby("Year", as_index=False)["Per Capita Energy Use (kWh)"].sum()
df_germany_grouped = df_germany.groupby("Year", as_index=False)["Per Capita Energy Use (kWh)"].sum()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 5))
plt.plot(df_canada_grouped["Year"], df_canada_grouped["Per Capita Energy Use (kWh)"], marker='o', linestyle='-', color='b', linewidth=2, markersize=6, label="Canadá")
plt.plot(df_germany_grouped["Year"], df_germany_grouped["Per Capita Energy Use (kWh)"], marker='s', linestyle='--', color='r', linewidth=2, markersize=6, label="Alemania")

# Personalizar el gráfico
plt.xlabel("Año", fontsize=12)
plt.ylabel("Consumo Per Cápita de Energía (kWh)", fontsize=12)
plt.title("Evolución del Consumo Per Cápita de Energía (2020-2025)", fontsize=14)
plt.xticks(df_canada_grouped["Year"], rotation=45)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar el consumo promedio por país
df_avg_consumption = df.groupby("Country", as_index=False)["Total Energy Consumption (TWh)"].mean()

# Ordenar de mayor a menor consumo
df_avg_consumption = df_avg_consumption.sort_values(by="Total Energy Consumption (TWh)", ascending=False)

# Generar colores únicos para cada país
num_paises = len(df_avg_consumption)
colores = plt.colormaps["tab20"](np.linspace(0, 1, num_paises))  # Solución actualizada

# Crear el gráfico de barras
plt.figure(figsize=(14, 7))
bars = plt.bar(df_avg_consumption["Country"], df_avg_consumption["Total Energy Consumption (TWh)"], color=colores)

# Personalizar el gráfico
plt.xlabel("País", fontsize=12)
plt.ylabel("Consumo Promedio de Energía (TWh)", fontsize=12)
plt.title("Consumo Promedio de Energía por País (Todos los Años)", fontsize=14)
plt.xticks(rotation=90, fontsize=10)  # Rotar etiquetas para mejor lectura
plt.yticks(fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.6)  # Agregar grilla horizontal

# Mostrar el valor de cada barra encima
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f"{yval:.1f}", ha="center", va="bottom", fontsize=9, color="black")

# Mostrar el gráfico
plt.show()
```
